```python
import requests
import json
import pprint as pp

import pandas as pd
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

from bs4 import BeautifulSoup

import datetime as dt
```

 ### This code reads from file and creates city's list

```python

#TODO: make sure that not need to chang the file path.
try:
    file = open(r'C:\Users\saadon100\Desktop\חומר לימודי\שנה ג\סמסטר ב\מבוא לידע ונתונים\מטלות\מטלה 5\destinations_LP_crawler_Ex5.csv')
#TODO: fix it
except:
    exit()

destinations_lines = file.read().split('\n')


# creates city and city_LP lists from the file
cities = []
cities_LP = []
for line in destinations_lines[1:]: #skips the first line
    city_format = line.split(',')

    #validates the line is not empty   
    if len(city_format) < 2:
        continue
        
    cities.append(city_format[0]) 
    cities_LP.append(city_format[1])
    
    
# print(cities_LP)    
```

### create county list from GEOLOCATION by given city







```python
#---------------ENTER HERE YOUR KEY--------------#
key_api = 'fictive'


countries = []
for city in cities:
# getting the name of the country from GOOGLE Geocoding

    url_geo = f"https://maps.googleapis.com/maps/api/geocode/json?address={city}&key="
    url_geo_full = url_geo+key_api

    payload={}
    headers = {}
    response_geo = requests.request("GET", url_geo_full, headers=headers, data=payload).json()

#     print(url_geo_full)
#     pp.pprint(response_geo)

#getting the first object in the list 'results'
    first_obj = response_geo['results'][0]
    country_list = first_obj['formatted_address'].split(',')[1:]
    
    # delete the wight spaces
    for inx in range(len(country_list)):
        country_list[inx] = country_list[inx].strip()
    countries.append(country_list)

# print(countries)



```

### collect all cities and countries in dict

```python
city_object = {}

for index in range(len(cities)):
    city_object[cities[index]] = {"city_lp": cities_LP[index],
                                  "country": countries[index]}
    
# pp.pprint(city_object)
```

### getting the html code and insert the text paragraph to the 

```python

for city, v in city_object.items():
#     print(city)
    city_lp = v['city_lp']
    country_list = v['country']
    for country in country_list:
        try:
            # make url and request html, 
            # if success - save the county, otherwise continue the next country.
            url = f'https://www.lonelyplanet.com/{country}/{city_lp}'
#             print(url)
            html = requests.get(url)
            if html.status_code == 200:
                city_object[city]["country"] = country
                
                #--------getting the PARAGRAPH TEXT from html.-------
                soup = BeautifulSoup(html.content,'html.parser')
                tag = soup.find("div", {"class": "readMore_content___5EuR relative overflow-hidden max-h-96 is-max wysiwyg"})

                text = ""
                for p in tag:
                    if p.name == 'h2':
                        text += p.get_text() + ': '
                    elif p.name == 'p':
                        text += p.get_text() + '\n'
                text = text[:-1] # delete the last space.
                #-----------------------------------------------------
                city_object[city]["text"] = text
                
#                 pp.pprint(city_object[city])   
                break # not try the next countries
        except Exception as e:
            print("ERROR: ")
            # nothing

            
# pp.pprint(city_object)         
    
 
#print("\nEND OF FILE")
```

 ### scrape the description of the destination in Dataframe and export to csv file

```python
city = []
country = []
description = []

for city_name in city_object.keys():
    city.append(city_name)
    country.append(city_object[city_name]['country'])
    description.append(city_object[city_name]['text'])

    

    
LP_destinations = pd.DataFrame([city,country,description]).T
LP_destinations.columns = ['city','country','description']

# print(LP_destinations)

LP_destinations.to_csv('LP_destinations.csv', index=False)
```
